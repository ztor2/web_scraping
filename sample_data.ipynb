```python
import json
import pandas as pd
```

```python
with open('product_sample.json', 'r', encoding='utf-8') as f:
    pdt = json.load(f, strict=False)
with open('cmt_sample.json', 'r', encoding='utf-8') as f:
    cmt = json.load(f, strict=False)
```

```python
product_data = pd.DataFrame(pdt)
product_data.dropna(axis=1, inplace=True)
product_data = product_data.loc[:, (product_data != 0).any(axis=0)]
product_data.iloc[:5, :14]
```

```python
cmt_data = pd.DataFrame(cmt)
cmt_data.dropna(axis=1, inplace=True)
cmt_data.iloc[:10]
```
